## **Definimos variables y funciones**

Importamos la librería **SymPy** de Python y definimos como symbolos la variable independiente $x$ y demas constantes y como función la variable dependiente $y$, como se aprecia a continuación:

```python
from sympy import *
from CalculusPy.SlopeFields import *

x = Symbol("x", real= True)
y = Function("y",real=True)
```

#   **CAMPOS DIRECCIONALES**

Un **campo direccional** o **campo de pendientes** es una representación visual de las soluciones de una ecuación diferencial ordinaria (EDO) en un plano.Está asociado con ecuaciones diferenciales de primer orden de la forma:

$$\frac{dy}{dx}=f(x,y)$$

Se dibujan pequeños segmentos de línea en distintos puntos del plano, cuyas pendientes corresponden a los valores de la función $f(x, y)$ en esos puntos. Estas lineas indican la dirección en la que cambia la variable dependiente ($y$) con respecto a la variable independiente ($x$).

La principal utilidad de un campo direccional es visualizar el comportamiento general de las soluciones de una EDO sin tener que resolverla explícitamente.

##    **Ejemplo 1: (Grafica del campo de pendientes)**
Grafica el campo de pendientes de la siguiente ecuación diferencial y dibuja a mano la curva solución aproximada que pasa por el punto indicado.

$$\frac{dy}{dx} = 2y-x , \quad y(0) = \frac{1}{2}$$


```python
dydx =2*y(x)-x

slope_fields(dydx,ics={y(0):1/2})
```

###    **Verificación solución particular:**

```python
ode = Eq(diff(y(x),x),2*y(x)-x)
sol = dsolve(ode,ics={y(0):1/2})
sol
```

```python
plot(sol.rhs,(x,-2,0.5),size=(4, 3))
```

##    **Ejemplo 2: (Grafica del campo de pendientes)**
Grafica el campo de pendientes de la siguiente ecuación diferencial y dibuja a mano la curva solución aproximada que pasa por el punto indicado.

$$\frac{dy}{dx} = x^2 y+x y , \quad y(0) = 1$$


```python
dydx = x**2*y(x)+y(x)*x

slope_fields(dydx,ics={y(0):1})
```

###    **Verificación solución particular:**

```python
ode = Eq(diff(y(x),x),x**2*y(x)+y(x)*x)
sol = dsolve(ode,ics={y(0):1})
sol
```

```python
plot(sol.rhs,(x,-1,1),size=(4, 3))
```
